```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Part A: Initial Exploration**

```python

# Load dataset (from ISLR package source)
url = "https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Wage.csv"
wage = pd.read_csv(url)

wage.head()
```

```python
#Wage vs Age

plt.figure(figsize=(8,5))
sns.scatterplot(x="age", y="wage", data=wage, alpha=0.5)
plt.title("Wage vs Age")
plt.xlabel("Age")
plt.ylabel("Wage")
plt.show()
```

The intended use of this scatter plot is to display the relationship between age and wage. From the plot we can conclude the older an employee gets the higher their wage is. This seems to not be a pattern once you get to the ages of 30 to 60 which have wages that range from $50k-$200k consistently. It is not particularly linear This suggests that other facts can have an influence on the flucation of wages.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="year", y="wage", data=wage)
plt.title("Wage vs Year")
plt.xlabel("Year")
plt.ylabel("Wage")
plt.show()
```

In this box plot, the wages appear to be stable and consistent across the years with outliers of wage increases during certain years. It displays its limited growth because the median does not really change at all. This graph is without an increase in wages over time


```python
plt.figure(figsize=(10,5))
sns.boxplot(x="education", y="wage", data=wage)
plt.xticks(rotation=45)
plt.title("Wage by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Wage")
plt.show()
```

This box plot intends to compare level of education to wages. We can conclude that wages and education level have a positive correlation with wages increasing as education level increaseing. The higher the education level completed, the higher the median age is for the group. For example, Advanced degree having the highest median. There are always outliers and varaibility within this data, indicating other factors could be influencing the relationship

**Part B: The Hedgehog View**

```python
wage.groupby("education")["wage"].mean()


# “If a person increases their education, their wages will increase because higher education is associated with higher wages (on average aka mean).”
```




My python groupby code compares education and wage using median shows that education is an extrmeely prominnent factor in the increase of wages. The relationships are consistent with High School grad have a lower average wage than college degrees--advanced degrees having the highest mean value.
